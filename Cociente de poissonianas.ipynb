```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
num_muestras = int(1e6)

xs = poisson.rvs(10, size=num_muestras)
ys = poisson.rvs(15, size=num_muestras)

zs = xs/ys
```

```python
#np.savetxt("muestra_cocientes_poisson.csv", zs, delimiter=",")
```

```python
#Elimino los infinitos
mask = (zs != np.float('+inf'))
zs = zs[mask]

# Verifico que para todos los elementos sea cierto que no son +inf
np.all(zs != np.float('+inf'))
```

```python
# Veo cu√°ntos elementos quedaron
len(zs)
```

```python
# Me copio del ploteo de Sami

fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(1, 1, 1)

bines = np.linspace(0,4,300)
numero, bins = np.histogram(zs, bins = bines)
error = (np.sqrt(numero) / (np.diff(bins) * np.sum(numero)))
numero = numero / (np.diff(bins) * np.sum(numero)) #normalizo el area
ax1.bar(bins[:-1], numero, width = np.diff(bins),  yerr=error, ecolor="b", color='dodgerblue', alpha=0.5)
#ax1.set_xlabel('k')
#ax1.set_ylabel('P(k)')
ax1.grid()
```
